REGRESION LINEAL

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("D:\lineal\data.csv",sep=",")
```

```python
data
```

```python
data.plot.scatter(x="metro",y="precio")
plt.show()
```

```python
# esta variable va a contenerla regresion lineal.
regresion = linear_model.LinearRegression()
```

```python
# Agregaremos los datos a un vector.
metro = data["metro"].values.reshape((-1,1))
```

```python
# Creamos el modelo.
modelo = regresion.fit(metro, data["precio"])

print("Interseccion (b)"), (modelo.intercept_)
```

```python
# imprimimos la pendiente.
print("Pendient (m)"), (modelo.coef_)
```

```python
#predicciones
entrada = [[15],[20]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
data.plot.scatter(x="metro", y="precio", label= 'Datos')
plt.scatter(entrada, predicciones, color='orange')
plt.plot(entrada, predicciones, color= 'green', label='Linea de Regresion')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
